```python
#Import Pandas Libraries
import pandas as pd
```

```python
# Load and Explore Data
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
# Input Data Preprocess 
Intdependent=dataset[["YearsExperience"]]
```

```python
Independent
```

```python
# Output Data Preprocess 
Dependent=dataset[["Salary"]]
```

```python
Dependent
```

```python
#Split Data into Training and Testing Sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(Independent,Dependent,test_size=0.30,random_state=0)
```

```python
X_train,X_test,Y_train,Y_test
```

```python
#Create the Train the Model
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,Y_train)
```

```python
# Checking Slope value 
weight=regressor.coef_
```

```python
weight
```

```python
# Checking Initial Value
Bios=regressor.intercept_
```

```python
Bios
```

```python
# Evaluate the Input test data
y_pred=regressor.predict(X_test)
```

```python
# Evaluate the Model
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,y_pred)
```

```python
r_score
```

```python
# File name creating to save
import pickle
filename="finalized_model_linear.sav"
```

```python
# Saving Model
pickle.dump(regressor,open(filename,'wb'))
```

```python
# Loading Saved Model
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
```
